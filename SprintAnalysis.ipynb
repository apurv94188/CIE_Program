```python
import pandas as pd

# Assign spreadsheet filename to `file`
file = 'AnalysisSheet.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

# Print the sheet names
print(xl.sheet_names)

# Load a sheet into a DataFrame by name: df1
df0 = xl.parse('Sheet1')
```

```python
df0.head(3)
```

```python
df1 = df0[df0["role"]==1].iloc[:,[10, 12, 13,15]]
df1
```

```python
df1 = df1.dropna()
```

```python
df1
```

```python
df1
```

```python
df1.reset_index(drop=True)
```

```python

df1["interviewed_customer"] = pd.to_numeric(df1['interviewed_customer'], errors='coerce')
df1["market_research"] = pd.to_numeric(df1['market_research'], errors='coerce')
#df1["market_research"] = pd.DataFrame(columns=[ 'market_research' ], dtype=float)
df1["research_analysis"] = df1["market_research"] + df1["interviewed_customer"]

df1
```

```python
df1["research_analysis"] = df1["research_analysis"]/df1["old_team"]
```

```python
df1[df1["Sprint"]==2.0]["research_analysis"]
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Bar(
    y = df1[df1["Sprint"]==1.0]["research_analysis"],
    x = df1[df1["Sprint"]==1.0]["Sprint"].tolist(),
    name='Attended Sprint'
)
trace2 = go.Bar(
    x = df1[df1["Sprint"]==2.0]["Sprint"].tolist(),
    y = df1[df1["Sprint"]==2.0]["research_analysis"],
    name='Not Attended Sprint Before'
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```
